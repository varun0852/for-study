```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import PromptTemplate , FewShotPromptTemplate
from langchain.messages import HumanMessage , SystemMessage , AIMessage
```

```python
llm =ChatGroq(
    model = "llama-3.1-8b-instant",
    api_key = "",
    temperature = 0.7, # when temperature is low model trying to be determinesting and wgen it high its trying to br probabilitisting
    max_tokens = None,
    timeout = None,
    max_retries=2,
)
```

## Zero Shot Prompting

```python
prompt = ChatPromptTemplate.from_messages([
    ("system" , "You are a senior AI Engineer"),
    ("human" , "Explain {topic} in one sentence")
])
```

```python
chain = prompt | llm | StrOutputParser()
```

```python
response = chain.invoke({"topic" : "LLM"})
```

```python
print(response)
```

## Few - Shot - Prompt

```python
examples = [
    {"input" : "Tell me about Generaative AI" , "output" : "Thanks for asking Varun, and Here is your answer to Generative AI Generative AI refers to a class of machine learning models that can create new, original content, such as images, videos, music, or text, that is often indistinguishable from human-created content, leveraging complex algorithms and neural networks to generate novel outputs."},
    {"input" : "Tell me about LLM" , "output" : "Thanks for asking Varun, and Here is your answer to LLM is A Large Language Model (LLM) is a type of artificial intelligence (AI) model that uses complex algorithms and massive amounts of training data to generate human-like language, answering questions, translating text, and even creating original content."}
]
```

```python
example_prompt = PromptTemplate(
    imput_variables = ["input" , "output"],
    template = "Input : {input}\nOutput : {output}"
)
```

```python
few_shot_prompt = FewShotPromptTemplate(
    examples = examples,
    example_prompt = example_prompt,
    prefix = "You are an Senior AI Manager in a Big Tech Company.\n",
    suffix = "\nInput: {user_input} \nOutput:",
    input_variables = ["user_input"]
)
```

```python
chain_few_shot = few_shot_prompt | llm | StrOutputParser()
```

```python
response = chain_few_shot.invoke({"user_input" : "What is gen AI"})
```

```python
print(response)
```

## Chain Of Thoughts Prompt

```python
chain_of_thought = """
you are a logical reasoning expert.

Question : {question}

Instruction:
1. Break down the question into small steps.
2. Solve each Step correctly.
3. Provide the final answer only after the reasoning.
4. if you don't know the answer just say i don't know don't make uo any answer

Let's think step by step:
"""
```

```python
prompt = ChatPromptTemplate.from_template(chain_of_thought)
```

```python
chain = prompt | llm | StrOutputParser()
```

```python
logical_question = """
Solve step by step:
I have 100 rupees.
If a pen costs 10 rupees and I buy 5 pens, and then i sell 2 pens
what is the total amount that i have after buying and selling the pens?
"""
```

```python
print("--- Chain of Thought Excuting ---")
response = chain.invoke({"question": logical_question})
print(response)
```

## Role play model

```python
messages = [
    {"role" : "system" , "content" : "you are a senior AI Engineer"},
    {"role" : "user" , "content" : "Explain LLM in one sentence"}
]
response = llm.invoke(messages)
print(response)
```

## Self Correction agent

```python
draft_prompt = ChatPromptTemplate.from_template(
#     """You are a knowledgeable technical instructor who helps students understand complex concepts politely accross a wide range of technical topics.
# Your expertise includes artificial intelligence, machine learning , Large Language Model (LLMs) , Neural Network , and programming in languages
# such as python , java , and more. You also provide in-depth support for AI engineering questions and other advanced techinical subjects.
# And You can also provide easy explination
# and if you don't know the answer just say i don't know don't make uo any answer"""
    "Draft a short polite email to a client declining their project request due to low budget: {input}"
)
draft_chain = draft_prompt | llm | StrOutputParser()
```

```python
draft_chain.invoke({"input" : "what is gen ai"})
```

```python
critique_prompt = ChatPromptTemplate.from_template(
    "Identify any tone issues or rudeness in the draft prompt. If it's perfect, say 'No issues'-\nDraft: {draft}"
)
critique_chain = critique_prompt | llm | StrOutputParser()
```

```python
refine_prompt = ChatPromptTemplate.from_template(
    "Rewrite the answer based on critique.\nOriginal Draft: {draft}\nCritique: {critique}\nFinal Email:"
)
refine_chain = refine_prompt | llm | StrOutputParser()
```

```python
chain = (
    {"draft" : draft_chain, "input" : RunnablePassthrough()}
    | RunnablePassthrough.assign(critique = critique_chain)
    | refine_chain
)
```

```python
print(chain.invoke("what is ai"))
```
