```python
# Import libraries
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Sample text data
documents = [
    "I love machine learning",
    "Machine learning is amazing",
    "I love coding in Python",
    "Python and machine learning are fun"
]

# ---------------------------
# 1. Bag of Words (BoW)
# ---------------------------
bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(documents)

print("=== Bag of Words ===")
print("Feature Names:", bow_vectorizer.get_feature_names_out())
print("BoW Matrix:\n", bow_matrix.toarray())

# ---------------------------
# 2. TF-IDF
# ---------------------------
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

print("\n=== TF-IDF ===")
print("Feature Names:", tfidf_vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
# Sample documents
documents = [
    "I love machine learning",
    "Machine learning is amazing",
    "I love coding in Python",
    "Python and machine learning are fun"
]

# Function to generate n-grams sequentially
def generate_ngrams(doc, n=2):
    tokens = doc.split()
    return [' '.join(tokens[i:i+n]) for i in range(len(tokens)-n+1)]

# Generate sequential 2-grams for each document
for i, doc in enumerate(documents):
    bigrams = generate_ngrams(doc, n=2)
    print(f"Document {i+1} 2-grams (sequential): {bigrams}")
```

```python
# Import library
from sklearn.feature_extraction.text import CountVectorizer

# Sample documents
documents = [
    "I love machine learning",
    "Machine learning is amazing",
    "I love coding in Python",
    "Python and machine learning are amazing and fun"
]

# Initialize CountVectorizer for BoW
bow_vectorizer = CountVectorizer()

# Fit and transform the documents
bow_matrix = bow_vectorizer.fit_transform(documents)

# Feature names (vocabulary)
feature_names = bow_vectorizer.get_feature_names_out()

# Display results
print("=== Bag of Words ===")
print("Feature Names:", feature_names)
print("BoW Matrix:\n", bow_matrix.toarray())
```
