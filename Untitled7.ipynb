```python
import nltk
from nltk.chat.util import Chat, reflections

pairs = [
    ["hi|hello", ["Hello!", "Hi there!"]],
    ["what is your name ?", ["I am a chatbot created using Python."]],
]

chat = Chat(pairs, reflections)
chat.converse()
```
