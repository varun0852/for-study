```python
# Population mean(Need to be provided by client)
pop_mean = 30

# Population Standard Deviation(Provided by client)
pop_std = 12.5

# sample mean
sample_mean = np.mean(data)

# length of sample data
n = len(data)

#Z_Score/Test which measures the difference in two population means
#when variance is known
# find the z-score = Number of Standard Deviation between Pop & Sample
z_score = (sample_mean - pop_mean)/(pop_std/np.sqrt(n))

# find the p - value
p_value = stats.norm.cdf(z_score)
# cdf = cumulative distribution function
print("The Z_Score value  is {} and the P_Value is {}".format(round(z_score , 4) , round(p_value , 4)))
```

We can see that the p-value is greater than 0.05, thus the null hypothesis is retained. Since 6.78% is greater than the significance value 5%, there is not enough evidence to reject null hypothesis. The null hypothesis is retained and it can be concluded that average delivery time is greater than or equal to 30.

when p_value < less than 0.05 then the probability of null hypothesis is 5%  and probability of alternate is 95% we accept alternate hypothesis  and if p_value is greter than 0.05 we accept null hypothesis
