# **LINEAR SCLAER**

The nature of the curve wont change it stays the same



# NORMALIZATION (MIN_MAX_SCALLING)
Scales the data in range of 0 to 1


the nature of the curve does not change

```python
min_max_scaler = MinMaxScaler() # the scale of the data will be between 0 to 1
social_media = min_max_scaler.fit_transform(df[["Social_Media"]])
print(min(social_media))
print(max(social_media))
sns.distplot(social_media)
```

## Standardization (Standard Scaling)
scales the in range of -2 to 2

the nature of the curve does not change

```python
df["Social_Media"]
```

```python
df["Product_Sold"]
```

```python
std_scale = StandardScaler()
df1 = std_scale.fit_transform(df[["Social_Media"]])
sns.distplot(df1)
```

# MAX ABS SCALLING
if we have negative data it will convert the range of the data between -1 to 1



```python
max_abs_scaler = MaxAbsScaler()
df2 = max_abs_scaler.fit_transform(df[["Social_Media"]])
sns.distplot(df2)
```

# ROBUST SCALER
it transform the data in such a way so that the outliers can be matched

```python
robust_scaler = RobustScaler()
df3 = robust_scaler.fit_transform(df[["Social_Media"]])
sns.distplot(df3)
```

# **POWER TRANSFORMER**

The nature of the curves can changes

# LOG TRANNSFORM

```python
x = np.log(df["Social_Media"])
sns.distplot(x)
```

# SQUARE ROOT TRANSFORMATION

```python
x1 = np.sqrt(df["Social_Media"])
sns.distplot(x1)
```

# RECIPROCAL TRANSFORMATION

```python
x3 = 1/df["Social_Media"]
sns.distplot(x3)
```

# **BOX COX TRANSFORMATION**
we have to provide a method = "box-cox" for it
it does not support negative value

```python
box_cox = PowerTransformer(method = "box-cox")
x4 = box_cox.fit_transform(df[["Social_Media"]])
sns.distplot(x4)
```

# **BOX COX TRANSFORMATION**(YEO JOHNSON TRANSFORMER )
If we dont provide method it will default take it as yeo johnson method


it support both negative and positive value

```python
box_cox_y = PowerTransformer()
x5 = box_cox_y.fit_transform(df[["Social_Media"]])
sns.distplot(x5)
```
