```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk

nltk.download("stopwords")
nltk.download("wordnet")

from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer , WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer , TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("spam.csv" , encoding="cp1252")
df
```

# CLEANING

```python
df1 = df[["v2" , "v1"]]
df1
```

```python
df1.isna().sum()
```

```python
df1[df1.duplicated()]
```

```python
df2 = df1.drop_duplicates(keep="first")
df2
```

```python
df2["v2"].isna().sum()
```

```python
# Converting into lower case
lower_case = []
for text in df2["v2"]:
  lower_case.append(text.lower())
```

```python
# Removing Punctuations
punctuation_no_removed = []
for text in lower_case:
  punctuation_no_removed.append(re.sub('[^a-zA-Z]' , ' ' , text))
```

```python
# Tokenizeding
tokenized = []
for text in punctuation_no_removed:
  tokenized.append(text.split())
```

```python
# Removing stopwords
parent_lst = []
for token_list in tokenized:
  stop_word_removed = []
  for word in token_list:
    if word not in stopwords.words("english"):
      stop_word_removed.append(word)
  parent_lst.append(stop_word_removed)
```

```python
lancaster = LancasterStemmer()
```

```python
lancaster_stemmed = []
for stop_word_removed in parent_lst:
  stemmed_token_lst = []
  for token in stop_word_removed:
    stemmed_token = lancaster.stem(token)
    stemmed_token_lst.append(stemmed_token)
  lancaster_stemmed.append(stemmed_token_lst)
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
wordnet_lemmatized = []
for stop_word_removed in parent_lst:
    lemmatized_token_lst = []
    for token in stop_word_removed:
        lemmatized_token = lemmatizer.lemmatize(token)
        lemmatized_token_lst.append(lemmatized_token)
    wordnet_lemmatized.append(lemmatized_token_lst)
```

```python
lancaster_stemmed_text = []
for stemmed_token_lst in lancaster_stemmed:
  text = ' '.join(stemmed_token_lst)
  lancaster_stemmed_text.append(text)
```

```python
wordnet_lemmatized_text = []
for lemmatized_token_lst in wordnet_lemmatized:
  text = ' '.join(lemmatized_token_lst)
  wordnet_lemmatized_text.append(text)
```

```python
df2["lancaster_stemmed_text"] = lancaster_stemmed_text
df2["wordnet_lemmatized_text"] = wordnet_lemmatized_text
```

```python
df2
```

```python
df2.isna().sum()
```

```python
def generate_ngrams(doc , n=2):
  tokens = doc.split()
  return [' '.join(tokens[i:i+n]) for i in range(len(tokens)-n+1)]

for i, doc in enumerate(df2["wordnet_lemmatized_text"]):
  bigrams = generate_ngrams(doc , n=2)
  print(f"Document {i+1} 2-grams (sequential): {bigrams}")
```

```python
vectorizer = CountVectorizer(ngram_range=(2,2))
bow = vectorizer.fit_transform(df2["wordnet_lemmatized_text"])
```

```python
vectorizer.get_feature_names_out()
```

```python
bow.toarray()[0].shape
```

```python
tfidf_vectorizer = TfidfVectorizer(ngram_range=(2,2))
tfidf = tfidf_vectorizer.fit_transform(df2["wordnet_lemmatized_text"])
```

```python
tfidf.toarray()[0].shape
```

```python
df3 = pd.DataFrame(tfidf.toarray(),columns = tfidf_vectorizer.get_feature_names_out())
```

```python
df2["v1"].isna().sum()
```

```python
df3 = df3.reset_index(drop=True)
label = df2["v1"].reset_index(drop=True)
```

```python
df3["label"] = label
```

```python
df3["label"].isna().sum()
```

```python
x = df3.drop(columns=["label"])
y = df3["label"]
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size=0.2 , random_state = 7)
```

```python
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(xtrain , ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
print(classification_report(ytest , ypred))
```

```python
confusion_matrix(ytest , ypred)
```
