```python
#Self-Correction Prompting / Reflection Prompting
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python
draft_prompt = ChatPromptTemplate.from_template(
    "Draft a short polite email to a client declining their project request due to low budget: {input}"
)
draft_chain = draft_prompt | llm | StrOutputParser() #Langchain Express Language
```

```python
draft_chain.invoke({"input": "What is gen ai"})
```

```python
critique_prompt = ChatPromptTemplate.from_template(
    "Identify any tone issues or rudeness in this email draft. If it's perfect, say 'No issues'.\nDraft: {draft}"
)
critique_chain = critique_prompt | llm | StrOutputParser()
```

```python
refine_prompt = ChatPromptTemplate.from_template(
    "Rewrite the email based on the critique.\nOriginal Draft: {draft}\nCritique: {critique}\nFinal Email:"
)
refine_chain = refine_prompt | llm | StrOutputParser()
```

```python
chain = (
    {"draft": draft_chain, "input": RunnablePassthrough()}
    | RunnablePassthrough.assign(critique=critique_chain)
    | refine_chain
)
```

```python
print(chain.invoke("The client offered $500 for a $5000 job."))
```

```python
#Majority Voting Prompting
```

```python
prompt = ChatPromptTemplate.from_template(
    "Solve this problem and give ONLY the best google search Keywords for AI Course: {question}"
)
```

```python
chain = prompt | llm | StrOutputParser()
```

```python
question = "As Im designing the google search ads, lot of money is getting wasted due to improper keywords. " \
"Give me the best keywords to make more money"
```

```python
responses = [chain.invoke({"question": question}) for _ in range(5)]
```

```python
print("--- All Responses ---")
for r in responses: print(r)
```

```python
from collections import Counter
final_answer = Counter(responses).most_common(1)[0][0]
```

```python
print(final_answer)
```

```python
#Few-Shot Prompt
```

```python
from langchain_core.prompts import PromptTemplate, FewShotPromptTemplate
```

```python
examples = [
    {"input": "Give me AI Roadmap", "output": "Thanks! For Asking Kanth, Here is your Roadmap learn with 1. Python, 2. SQL, 3. Ml, 4. Deep Learning, 5. Gen AI, 6. Deployment and Work on Real-Time PRojects"},
    {"input": "Give me Data Analytics Roadmap", "output": "Thanks! For Asking Kanth, Here is your Roadmap learn with 1. Power BI, 2. Tableau, 3. Data Warehouse using SQL, 4. Python and work on Real-Time Projects"},
]
```

```python
example_prompt = PromptTemplate(
    input_variables=["input", "output"],
    template="Input: {input}\nOutput: {output}"
)
```

```python
few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix="You are an Head of AI in Big Tech Company.\n",
    suffix="\nInput: {user_input}\nOutput:",
    input_variables=["user_input"]
)
```

```python
chain_few_shot = few_shot_prompt | llm | StrOutputParser()
```

```python
response = chain_few_shot.invoke({"user_input": "Data Engineer Roadmap"})
```

```python
print(response)
```

```python
#chain of though Prompt
```

```python
cot_template = """
You are a logical reasoning expert.

Question: {question}

Instruction:
1. Break down the problem into small steps.
2. Solve each step explicitly.
3. Provide the final answer only after the reasoning.

Let's think step by step:
"""
```

```python
prompt = ChatPromptTemplate.from_template(cot_template)
```

```python
chain = prompt | llm | StrOutputParser()
```

```python
logic_question = """
I have 3 apples. I eat 2. Then I buy 3 more.
I give 1 to my brother. How many apples do I have now?
"""
```

```python
print("--- Chain of Thought Execution ---")
response = chain.invoke({"question": logic_question})
print(response)
```
