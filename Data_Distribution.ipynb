```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import math
```

```python
sns.set(style="whitegrid")
```

```python
# Function to plot a distribution and print skewness and kurtosis
def plot_distribution(data, title):
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)
    print(f"{title} - Skewness: {skewness:.2f}, Kurtosis: {kurtosis:.2f}")

    plt.figure(figsize=(8, 5))
    sns.histplot(data, kde=True, bins=30, color="blue", edgecolor="black", alpha=0.7)
    plt.title(f"{title}\nSkewness: {skewness:.2f}, Kurtosis: {kurtosis:.2f}", fontsize=14)
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()
```

```python
def normal_distribution(x, mu, sigma):
    """Calculate the PDF of the normal distribution at x."""
    coefficient = 1 / (sigma * math.sqrt(2 * math.pi))
    exponent = -((x - mu) ** 2) / (2 * sigma ** 2)
    return coefficient * math.exp(exponent)
```

```python
normal_data = np.random.normal(loc=0, scale=1, size=10000)
mean = normal_data.mean()
std = normal_data.std()

lst_prob = []
for i in normal_data:
    lst_prob.append(normal_distribution(i, mean, std))
```

```python
cdf = []
x = 0
for i in lst_prob:
    x+=i
    cdf.append(x)
```

```python
# 1️ Normal Distribution (Gaussian)
normal_data = np.random.normal(loc=0, scale=1, size=10000)
plot_distribution(normal_data, "Normal Distribution")
```

```python
# 2️ Uniform Distribution
uniform_data = np.random.uniform(low=-1, high=1, size=10000)
plot_distribution(uniform_data, "Uniform Distribution")
```

```python
# 3️ Right-Skewed Distribution (Exponential)
exp_data = np.random.exponential(scale=1, size=10000)
plot_distribution(exp_data, "Exponential Distribution (Right-Skewed)")
```

```python
# 4️ Left-Skewed Distribution (Transformed Exponential)
left_skewed_data = -np.random.exponential(scale=1, size=10000)
plot_distribution(left_skewed_data, "Left-Skewed Distribution")
```

```python
# 5 Leptokurtic Distribution (Heavy tails, High peak)
# Using Laplace distribution to create a leptokurtic distribution
leptokurtic_data = np.random.laplace(loc=0, scale=1, size=10000)
plot_distribution(leptokurtic_data, "Leptokurtic Distribution (Laplace)")
```

```python
# 6 Mesokurtic Distribution (Normal distribution with kurtosis 3)
# Normal distribution (mesokurtic)
mesokurtic_data = np.random.normal(loc=0, scale=1, size=10000)
plot_distribution(mesokurtic_data, "Mesokurtic Distribution (Normal)")
```

```python
# 7 Platykurtic-like Distribution (Using Narrower Normal Distribution for Light Tails)
# We can simulate platykurtic behavior by narrowing the normal distribution (small variance)
platykurtic_normal_data = np.random.normal(loc=0, scale=3, size=10000)  # Small variance

plot_distribution(platykurtic_normal_data, "Platykurtic-like Normal Distribution (Narrowed)")
```

```python
# 8 Bimodal Distribution (Combination of Two Normals)
bimodal_data = np.concatenate([np.random.normal(-2, 1, 5000), np.random.normal(3, 1, 5000)])
plot_distribution(bimodal_data, "Bimodal Distribution")
```

```python
# 9 Poisson Distribution
poisson_data = np.random.poisson(lam=5, size=10000)
plot_distribution(poisson_data, "Poisson Distribution")
```

```python
# 9 Binomial Distribution
binomial_data = np.random.binomial(n=10, p=0.5, size=10000)
plot_distribution(binomial_data, "Binomial Distribution")
```

```python
# 10 Bernoulli Distribution
bernoulli_data = np.random.binomial(n=1, p=0.5, size=10000)  # n=1 makes it Bernoulli
plot_distribution(bernoulli_data, "Bernoulli Distribution")
```

```python
# 11 Log-Normal Distribution
lognormal_data = np.random.lognormal(mean=0, sigma=1, size=10000)
plot_distribution(lognormal_data, "Log-Normal Distribution")
```

```python
# 12 Power Law Distribution (Pareto)
pareto_data = (np.random.pareto(a=3, size=10000) + 1)  # Shifted to make values positive
plot_distribution(pareto_data, "Power Law Distribution (Pareto)")
```
